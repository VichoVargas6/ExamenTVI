```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Abrir primer archivo
df = pd.read_csv('ArchivoMatr_Adm2022.csv', delimiter=';')
```

```python
df.head
```

```python
#Abrir segundo archivo
df1 = pd.read_csv('ArchivoD_Adm2022.csv', delimiter=';')
```

```python
df1.head
```

```python
#Abrir tercer archivo
df2 = pd.read_csv('ArchivoB_Adm2022.csv', delimiter=';')
df2
```

```python
#Juntamos los 3 datasets en la columna 'ID_aux'
datos = pd.merge(pd.merge(df, df1, on='ID_aux'), df2, left_on='ID_aux', right_on='ID_aux')
```

```python
#Creamos el nuevo dataset con las columnas de interes
new_col = ['ID_aux', 'PTJE_POND', 'SITUACION_POSTULANTE', 'INGRESO_PERCAPITA_GRUPO_FA']
datos1 = datos[new_col]
```

```python
datos1.head
```

```python
#Convertir 'PTJE_POND' a string
if datos1['PTJE_POND'].dtype != 'object':
    datos1.loc[:, 'PTJE_POND'] = datos1['PTJE_POND'].astype(str)

#Convertir 'PTJE_POND a float
datos1.loc[:, 'PTJE_POND'] = datos1['PTJE_POND'].str.replace(',', '.').astype(float)

#Convertir 'INGRESO_PERCAPITA_GRUPO_FA' a float
datos1.loc[:, 'INGRESO_PERCAPITA_GRUPO_FA'] = datos1['INGRESO_PERCAPITA_GRUPO_FA'].astype(float)
```

```python
datos1
```

```python
#Tomar solo los deciles del 1 al 10, ya que 99 es que no se mencionó
datos2 = datos1[datos1['INGRESO_PERCAPITA_GRUPO_FA'].between(1,10)]
```

```python
#Crear y mostrar el BoxPlot
plt.figure(figsize=(10, 6))
sns.boxplot(data=datos2, x='INGRESO_PERCAPITA_GRUPO_FA', y='PTJE_POND')
plt.title('Distribución del Puntaje segun Ingreso Per Cápita PAES 2022')
plt.xlabel('Decil Ingreso Per Cápita')
plt.ylabel('Puntaje Ponderado')
plt.show()
```
